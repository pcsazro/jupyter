```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
%matplotlib inline
import platform
path = "c:/Windows/Fonts/malgun.ttf"
from matplotlib import font_manager, rc
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~') 
import seaborn as sns
```

```python
shopping = pd.read_excel('final_project_data/2018~2020.04 온라인쇼핑 가구 판매액(억).xlsx')
shopping
```

```python
label = shopping['기간']
index = np.arange(len(label))
```

```python
plt.figure(figsize=(10,10))
plt.bar(index, shopping['2018년'])
plt.title('가구 판매액(억)', fontsize=20)
plt.xlabel('Month', fontsize=18)
plt.ylabel('거래액(억)', fontsize=18)
plt.xticks(index, label, fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,10))
bar_width = 0.2
alpha = 0.9
p1 = plt.bar(index, shopping['2018년'], 
             bar_width, 
             color='b', 
             alpha=alpha,
             label='2018')
p2 = plt.bar(index + bar_width, shopping['2019년'], 
             bar_width, 
             color='r', 
             alpha=alpha,
             label='2019')
p3 = plt.bar(index + bar_width + bar_width, shopping['2020년'], 
             bar_width, 
             color='g', 
             alpha=alpha,
             label='2020')
plt.title('가구 거래액(억)', fontsize=20)

plt.ylabel('기간', fontsize=18)

plt.xlabel('거래액(억)', fontsize=18)

plt.xticks(index, label, fontsize=15)

plt.legend((p1[0], p2[0], p3[0]), ('2018', '2019', '2020'), fontsize=15)

plt.show()
```

```python
age = pd.read_excel('final_project_data/서울시 1인가구(연령별).xlsx')
```

```python
age = age.set_index('연령')
```

```python
age
```

```python
plt.figure(figsize=(20,10))

age[['남자','여자']].plot(kind='bar')

plt.xticks(rotation=45)

plt.show()
```

```python
label = age['연령']
index = np.arange(len(label))
```

```python
plt.figure(figsize=(15,10))
bar_width = 0.2
alpha = 0.9
p1 = plt.bar(index, age['남자'], 
             bar_width, 
             color='b', 
             alpha=alpha,
             label='2018')
p2 = plt.bar(index + bar_width, age['여자'], 
             bar_width, 
             color='r', 
             alpha=alpha,
             label='2019')

plt.title('1인 가구 거주수(명)', fontsize=20)

plt.ylabel('인구수', fontsize=18)

plt.xlabel('연령', fontsize=18)

plt.xticks(index, label, fontsize=15)

plt.legend((p1[0], p2[0]), ('남자', '여자'), fontsize=15)

plt.show()
```

```python
live = pd.read_excel('final_project_data/서울시 1인가구(거처종류별).xlsx')
```

```python
live = live.set_index(live['성별'])
```

```python
live = live.T
```

```python
live
```

```python
live = live.drop(['성별'])
```

```python
live = live.reset_index()
```

```python
live
```

```python
label = live['index']
index = np.arange(len(label))
```

```python
plt.figure(figsize=(15,10))
bar_width = 0.2
alpha = 0.9
p1 = plt.bar(index, live['남자'], 
             bar_width, 
             color='b', 
             alpha=alpha,
             label='남자')
p2 = plt.bar(index + bar_width, live['여자'], 
             bar_width, 
             color='r', 
             alpha=alpha,
             label='여자')
p3 = plt.bar(index + bar_width + bar_width, live['계'], 
             bar_width, 
             color='g', 
             alpha=alpha,
             label='계')
plt.title('주거지별 1인가구수', fontsize=20)

plt.ylabel('인구수(명)', fontsize=18)

plt.xlabel('거주 종류', fontsize=18)

plt.xticks(index, label, fontsize=15, rotation=45)

plt.legend((p1[0], p2[0], p3[0]), ('남자', '여자', '계'), fontsize=15)

plt.show()
```
