# 3)

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

train - test ë¶„ë¥˜

```python
y = normalization_wine.loc[:,'quality']
y
```

```python
x = normalization_wine.loc[:,['fixed acidity','volatile acidity','citric acid','total sulfur dioxide','density','sulphates','alcohol']]
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, train_size=0.9, shuffle=True)
```

KNN

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train, y_train)
```

ì •ë‹µë¥  ê³„ì‚°

```python
y_pred = knn.predict(x_test)
```

```python
print('ì •ë‹µë¥  : ', accuracy_score(y_test, y_pred))
```

```python
classification_report(y_test, y_pred)
```

```python
test_wine = [10.,1.,0.5,50.,1.,0.5,10.]
```

```python
knn.predict_proba([test_wine])
```
